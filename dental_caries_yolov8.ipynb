```python
from ultralytics import YOLO

model = YOLO("yolov8m.pt")

model.train(
    data="/kaggle/working/dataset/data.yaml",
    epochs=150,
    imgsz=1024,
    batch=12,
    optimizer="AdamW",
    lr0=0.001,
    cos_lr=True,
    patience=0,
    mosaic=1.0,
    close_mosaic=10,
    project="dental",
    name="caries_final"
)
```
