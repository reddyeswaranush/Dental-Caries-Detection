```python
%%writefile /kaggle/working/dataset/data.yaml
train: train/images
val: valid/images
test: test/images

nc: 1
names:
  0: caries
```

```python
%%bash
LABEL_DIR="/kaggle/working/dataset"

for split in train valid test; do
  if [ -d "$LABEL_DIR/$split/labels" ]; then
    sed -i 's/^1 /0 /g' $LABEL_DIR/$split/labels/*.txt
  fi
done
```

```python
from ultralytics import YOLO

model = YOLO("yolov8m.pt")

model.train(
    data="/kaggle/working/dataset/data.yaml",
    epochs=180,
    imgsz=1024,
    batch=8,
    optimizer="AdamW",
    lr0=0.001,
    cos_lr=True,
    patience=0,
    mosaic=1.0,
    close_mosaic=10,
    project="dental",
    name="caries_final"
)
```
